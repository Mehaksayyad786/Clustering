# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
X=df[['midterm_marks','final_marks','quiz1_marks','quiz2_marks','quiz3_marks','previous_gpa','total_lectures','lectures_attended','total_lab_sessions','labs_attended']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_pca=pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
X_pca
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4,random_state=42)
```

```python
clusters=kmeans.fit_predict(X_pca)
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=clusters)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='pink',marker="X",s=250,label='centroids')
plt.xlabel("Principal component 1")
plt.ylabel("Principal component 2")
plt.title("Principal component Analysis with KMeans")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
score=silhouette_score(X_pca,clusters)
```

```python
score
```
